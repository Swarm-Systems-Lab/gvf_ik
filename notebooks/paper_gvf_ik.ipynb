```python
# Import standard libraries
import os

# Third-Party Libraries
import numpy as np
import matplotlib.pyplot as plt

# Import the Swarm Systems Lab Simulator
from ssl_simulator.visualization import set_paper_parameters, fixedwing_patch
set_paper_parameters(fontsize=15)

# -------------------------------------------------------------------------------------
# Define constants for file paths
OUTPUT_DIR = os.path.join("..", "output")
SIMDATA_FILE = os.path.join(OUTPUT_DIR, "data_paper.csv")
```

# Auxiliary figures for the paper/poster

```python
# Define the vectors
origin = [0, 0]
vector1 = [0, 2]
vector2 = [-1, 0]
vector3 = [2, 0] 
final_vector = [vector1[0] + vector2[0] + vector3[0], vector1[1] + vector2[1] + vector3[1]]  # Final vector

# Plot the vectors
fig, ax = plt.subplots(dpi=100, figsize=(8, 8))
ax.quiver(*origin, *vector1, angles="xy", scale_units="xy", scale=1, color="darkgreen", zorder=3)
ax.quiver(*origin, *vector2, angles="xy", scale_units="xy", scale=1, color="darkred", zorder=3)
ax.quiver(*origin, *vector3, angles="xy", scale_units="xy", scale=1, color="b", zorder=3)
ax.quiver(*origin, *final_vector, angles="xy", scale_units="xy", scale=1, color="grey", linewidth=2)
ax.text(vector1[0] / 2 + 0.1, vector1[1] - 0.2, r"$v_T$", fontsize=24, color="darkgreen")
ax.text(vector2[0] + 0.1, vector2[1] + 0.2, r"$v_B$", fontsize=24, color="darkred")
ax.text(vector3[0]*0.8, vector3[1] + 0.2, r"$v_C$", fontsize=24, color="b")
ax.text(final_vector[0] + 0.1, final_vector[1] - 0.2, r"$f$", fontsize=24, color="grey")

icon = fixedwing_patch([0,0], np.pi/2, 1, fc="royalblue", ec="black", zorder=5)
ax.add_patch(icon)  

# Set plot limits and labels
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.axhline(0, color="black", linewidth=0.5)
ax.axvline(0, color="black", linewidth=0.5)
ax.grid(color = "gray", linestyle = "--", linewidth = 0.5)
ax.set_aspect("equal", adjustable="box")
plt.show()
```
